```python
import networkx as nx
import matplotlib.pyplot as plt

# Graph erstellen
G = nx.DiGraph()
G.add_edges_from([
    ("Client 3", "Bluefield 3"),
    ("Client 2", "Bluefield 3"),
    ("Client 1", "Bluefield 3"),
    ("Bluefield 3", "Backend 1"),
    ("Bluefield 3", "Backend 2"),
    ("Bluefield 3", "Backend 3")
])

# Manuelle Positionen definieren
pos = {
    "Client 1": (0.2, 1.0),
    "Client 2": (0.5, 1.0),
    "Client 3": (0.8, 1.0),
    "Bluefield 3": (0.5, 0.6),
    "Backend 1": (0.2, 0.2),
    "Backend 2": (0.5, 0.2),
    "Backend 3": (0.8, 0.2),
}

# Kantenbeschriftungen
edge_labels = {
    ("Client 1", "Bluefield 3"): "Client Request",
    ("Bluefield 3", "Backend 1"): "Data 1",
    ("Bluefield 3", "Backend 2"): "Data 2",
    ("Bluefield 3", "Backend 3"): "Data 3",
}

# Graph zeichnen
plt.figure(figsize=(12, 8))
nx.draw(
    G, pos, with_labels=True,
    node_size=6000, node_color="skyblue",
    font_size=12, font_weight="bold",
    arrowsize=20, arrowstyle="->"
)

# Kantenbeschriftungen hinzufgen
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=edge_labels, 
    font_size=10, font_color="red"
)

# Titel hinzufgen
plt.title("Bluefield 3 Load Balancing", fontsize=16)

# Beschreibungstext unterhalb des Graphen
description = (
    "Dieser Graph zeigt eine typische Load-Balancing-Architektur:\n"
    "- Mehrere Clients senden Anfragen an einen Bluefield 3 Load Balancer.\n"
    "- Der Load Balancer verteilt die Anfragen gleichmig auf mehrere Backends.\n"
    "Dies sorgt fr Lastverteilung und Ausfallsicherheit."
)
plt.text(
    0.55, 0.6, "Die Bluefield 3 NIC ist in einem Hostsystem verbaut.\nAuf dieses Hostsystem zeigt die virtuelle IP Adresse."
)
plt.text(
    0.5, -0.1, description, fontsize=10, ha="center", va="top", wrap=True, transform=plt.gca().transAxes
)

# Graph anzeigen
plt.show()
```
