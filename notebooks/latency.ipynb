```python
def delta(data):
    return [(i[0] - i[1])/1000 for i in data]

def get_average(data):
    mean = 0
    for i in data:
        mean += i
    return mean/len(data)

baseline_data = [(421869, 419882), (469866, 467905), (517881, 515884), (565937, 563924), (25946, 23943)]
baseline_data = delta(baseline_data)
print(baseline_data)

thirtie_mio_data = [(702260, 699988), (654355, 652155), (606352, 603938), (558404, 555997), (510380, 507992)]
thirtie_mio_data = delta(thirtie_mio_data)
print(thirtie_mio_data)

ninety_mio_data = [(646610, 643964), (597800, 595953), (550587, 547995), (346418, 343924), (298200, 295916)]
ninety_mio_data = delta(ninety_mio_data)
print(ninety_mio_data)

baseline_mean = get_average(baseline_data)
thirtie_mio_mean = get_average(thirtie_mio_data)
ninety_mio_mean = get_average(ninety_mio_data)

print(baseline_mean)
print(thirtie_mio_mean)
print(ninety_mio_mean)
```

```python
import matplotlib.pyplot as plt

x = [0, 30000000, 90000000]
y = [baseline_mean, thirtie_mio_mean, ninety_mio_mean]
plt.figure(figsize=(10, 6))
plt.plot(x, y, "-s", label="RTT Latency")
plt.title("Relationship Between LB Load and RTT")
plt.xlabel("Packets/s Baseload")
plt.xticks(x, [str(int(val/1_000_000)) + " Mio." for val in x], rotation=45)
plt.ylabel("Roundtrip Time in ms")
plt.legend()
plt.grid(True)
plt.show()

```
